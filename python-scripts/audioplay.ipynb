```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display

# Load audio data from .npy file
audio_data = np.load("raw_i2s.npy")
plt.plot(audio_data)

# Sampling rate and duration of the audio
sampling_rate = 8000  # Replace with the actual sampling rate
duration = len(audio_data) / sampling_rate

# Display audio using IPython Display
display(Audio(data=audio_data, rate=sampling_rate))
print(f"Audio Duration: {duration:.2f} seconds")
```
